```python
#loading the data set that we are working on 
with open("the-verdict.txt","r",encoding="utf-8") as f:
    raw_text= f.read()
print("No. of characters: ",len(raw_text))
print(raw_text[:99])
```

```python
# using re- regex library to split between words .
import re
# some special characters as declared here will be treated as indiviual tokens
preprocessed = re.split(r'([,.:;?_!"()\']|--|\s)', raw_text)
preprocessed = [item.strip() for item in preprocessed if item.strip()]
print(preprocessed[:30])
```

```python
# to convert the text tokens to numerical token Ids we will build a vocabulary 
# a vocab is dictionary of all the words in the dataset and each word is mapped to an ID
# this dictionary will be used when we encode a given text as a lookup table
all_words = sorted(set(preprocessed))
vocab = {token:integer for integer,token in enumerate(all_words)}
for i,item in enumerate(vocab.items()):
    print(item)
    if i>=50:
        break
```

```python
# now what if we encode a text and a word present in it is not a part of our models vocab
# to encounter this problem and make sure there are no errors a very large dataset will be used to make the vocab
# but to make sure there are no errors will make new tokens for unkown words - <|unk|>
# when two different texts are given to a model a join token is used to separate them - <|endoftext|>
all_words.extend(["<|unk|>","<|endoftext|>"])
vocab = {token:integer for integer,token in enumerate(all_words)}
```

```python
# building the tokeniser class : encode+decode methods
class SimpleTokenizer:
    def __init__(self, vocab):
        self.str_to_int = vocab
        self.int_to_str = { i:s for s,i in vocab.items()}
    
    def encode(self, text):
        preprocessed = re.split(r'([,.:;?_!"()\']|--|\s)', text)
        preprocessed = [item.strip() for item in preprocessed if item.strip()]
        preprocessed = [
            item if item in self.str_to_int 
            else "<|unk|>" for item in preprocessed
        ]

        ids = [self.str_to_int[s] for s in preprocessed]
        return ids
        
    def decode(self, ids):
        text = " ".join([self.int_to_str[i] for i in ids])
        # Replace spaces before the specified punctuations
        text = re.sub(r'\s+([,.:;?!"()\'])', r'\1', text)
        return text
```

```python
tokenizer = SimpleTokenizer(vocab)

text1 = "Hello, do you like tea?"
text2 = "In the sunlit terraces of the palace."

text = " <|endoftext|> ".join((text1, text2))

print(text)
```

```python
tokenizer.encode(text)
```

```python
tokenizer.decode(tokenizer.encode(text))
```
